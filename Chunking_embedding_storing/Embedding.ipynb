```python
import os
def get_file_paths(folder_path):
    text_file_paths = []
    for root,dirs,files in os.walk(folder_path):
        for file in files:
            if file.endswith(".txt"):
                file_path = os.path.join(root, file)
                # Ensure double backslashes in paths for Windows
                text_file_paths.append(file_path.replace('\\', '/'))
    return text_file_paths

folder_path = '../Knowledgebase'
text_files = get_file_paths(folder_path)
```

```python

from langchain.embeddings.base import Embeddings
from sentence_transformers import SentenceTransformer

class SentenceTransformerEmbeddingModel(Embeddings):
    def __init__(self,modelname : str):
        self.model = SentenceTransformer(modelname)
    
    def embed_documents(self, texts):
        return [self.model.encode(d).tolist() for d in texts]

    def embed_query(self, query):
        return self.model.encode([query])[0].tolist()


```

```python
model = 'sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2'

myembedingmodel = SentenceTransformerEmbeddingModel(model)
```

```python
import faiss
from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_community.vectorstores import FAISS
```

```python
index = faiss.IndexFlatL2(len(myembedingmodel.embed_query("hello world")))

vector_store = FAISS(
    embedding_function=myembedingmodel,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={},
)
```

```python
from langchain_core.documents import Document
from uuid import uuid4

docs = []
for file in text_files:
    text_chunks = ovelap_chunking(file,200)
    for chunk in text_chunks:
        print(chunk)
        doc = Document(page_content=chunk.page_content,metadata={"source":f"{file}"})
        docs.append(doc)
```

```python
# Generate a list of unique UUIDs for each document in `docs`
uuids = [str(uuid4()) for _ in range(len(docs))]

# Add the documents to the vector store with the generated UUIDs
vector_store.add_documents(documents=docs, ids=uuids)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Initialize the model
model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')

def precision_at_k(actual_response, expected_response, threshold=0.8):
    """
    Calculate Precision@K based on cosine similarity.
    
    Parameters:
    - actual_response (list): List of strings (actual responses).
    - expected_response (list): List of strings (expected responses).
    - threshold (float): Cosine similarity threshold for considering a match.
    
    Returns:
    - float: Precision@K value.
    """
    a_emb = [model.encode(resp.page_content) for resp in actual_response]
    e_emb = [model.encode(resp) for resp in expected_response]
    
    tp = 0  # True positives
    k = len(actual_response)
    
    for a_vec in a_emb:
        # Calculate similarity of actual response to all expected responses
        similarities = [cosine_similarity([a_vec], [e_vec])[0][0] for e_vec in e_emb]
        
        # Check if any similarity exceeds the threshold
        if any(sim >= threshold for sim in similarities):
            tp += 1
    
    # Precision@K = True Positives / K
    precision = tp / k if k > 0 else 0
    return precision

#recal@k

```

```python
presition = 0
k_val = 0
for i in retrival_eval:
    result = vector_store.similarity_search(i.get("query"),k = 2)
    # eval = retrival_evalution()
    presistion = precision_at_k(actual_response=result,expected_response=i.get("relevant_docs"))
    print(presistion)
    k_val += 1
    presition += presistion
    # break

presistion,k_val
```
